```python
# Import necessary libraries for data analysis and cleaning
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

# Clean Drug

```python
# Load the drug dataset
df_drug = pd.read_csv('stratified_drug_200_withText_first.csv')

# Display basic information about the dataset
print("Dataset Shape:", df_drug.shape)
print("\nFirst few rows:")
df_drug.head()
```

```python
# Sort the DataFrame by 'respondent_ID'
df_drug.sort_values(by='Respondent_ID', inplace=True)

# Reset the 'Respondent_ID' column to be a sequence from 1 to 200
df_drug['Respondent_ID'] = np.arange(1, len(df_drug) + 1)

# Display the first few rows to verify the changes
df_drug.head()
```

```python
# Identify columns that contain 'Age' in their name
age_columns = [col for col in df_drug.columns if 'Age' in col]

# Round the values in these columns to 0 decimal places
for col in age_columns:
    df_drug[col] = df_drug[col].round(0)

# Display the first few rows and data types to verify the changes
print("Data types of age-related columns after rounding:")
print(df_drug[age_columns].dtypes)
print("\nFirst few rows with rounded age values:")
df_drug.head()
```

```python
# Check for NaN values in the columns
print("NaNs in Q1_Age:", df_drug['Q1_Age'].isnull().sum())
print("NaNs in Q15_FirstUseAge:", df_drug['Q15_FirstUseAge'].isnull().sum())

# Find rows where the condition is False, which indicates a potential data error
inconsistent_data = df_drug[~(df_drug['Q1_Age'] >= df_drug['Q15_FirstUseAge'])]

print("\nRows with inconsistent age data:")
inconsistent_data[['Q1_Age', 'Q15_FirstUseAge']]
```

```python
# Define the education column and the mapping
education_mapping = {
    1: 'Did not complete Year 12',
    2: 'Completed Year 12 (VCE)',
    3: 'TAFE Qualification',
    4: 'Undergraduate University Degree',
    5: 'Postgraduate University Degree'
}

df_drug['Q3_Education'] = df_drug['Q3_Education'].map(education_mapping)
# Display the first few rows of the modified columns to verify the changes
df_drug[['Q3_Education']].head()
```

```python
# Define the Likert scale columns and the mapping
likert_columns = [
    'Q8.1_Stress', 'Q8.2_PeerPressure', 'Q8.3_Normality', 
    'Q8.4_ExploreIdentity', 'Q8.5_HomeProblem', 'Q8.6_ConfidentToSayNo', 
    'Q9_Beliefs', 'Q10_MoralFailing', 'Q11_Penalties'
]

likert_mapping = {
    1: 'Strongly Disagree',
    2: 'Disagree',
    3: 'Neutral',
    4: 'Agree',
    5: 'Strongly Agree'
}

# Round and map the values for each Likert scale column
for col in likert_columns:
    # Ensure the column exists before processing
    if col in df_drug.columns:
        df_drug[col] = df_drug[col].round(0).map(likert_mapping)

# Display the first few rows of the modified columns to verify the changes
df_drug[likert_columns].head()
```

```python
df_drug.head()
```

```python
# Define the effectiveness scale columns and the mapping
effectiveness_columns = [
    'Q21.1_SchoolEd', 'Q21.2_ParentalGuide', 'Q21.3_Commu', 
    'Q21.4_PublicCamp', 'Q21.5_MentalHealth'
]

effectiveness_mapping = {
    1: 'Not effective at all',
    2: 'Not effective',
    3: 'Neutral',
    4: 'Effective',
    5: 'Very effective'
}

# Round and map the values for each effectiveness scale column
for col in effectiveness_columns:
    # Ensure the column exists before processing
    if col in df_drug.columns:
        df_drug[col] = df_drug[col].round(0).map(effectiveness_mapping)

# Display the first few rows of the modified columns to verify the changes
df_drug[effectiveness_columns].head()
```

```python
df_drug.head()
```

```python
# Define the condition where 'Q14_EverUsed' is 'No'
condition = df_drug['Q14_EverUsed'] == 'No'

# Identify columns for questions 15 through 20
# We can do this by checking if the column name starts with the relevant question numbers
columns_to_clear = [col for col in df_drug.columns if col.startswith(
    ('Q15', 'Q16', 'Q17', 'Q18', 'Q19', 'Q20')
)]

# For rows where the condition is met, set the values in the identified columns to NaN
df_drug.loc[condition, columns_to_clear] = np.nan

# Display a sample of rows where 'Q14_EverUsed' is 'No' to verify the changes
print("Sample of rows where 'Q14_EverUsed' is 'No' after clearing subsequent answers:")
df_drug[df_drug['Q14_EverUsed'] == 'No'].head()
```
